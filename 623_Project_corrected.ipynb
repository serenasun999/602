```python
import pandas as pd
import numpy as np
```

```python
IST_org = pd.read_csv("https://raw.githubusercontent.com/serenasun999/602_601Projects/data/IST_corrected.csv", encoding='latin-1')
IST_org
```

```python
IST_filtered = IST_org.copy()
IST_filtered[['DDIAGISC', 'DDIAGHA', 'DDIAGUN']]
```

```python
type_list = []

for index, row in IST_filtered.iterrows():
    if sum(row[['DDIAGISC', 'DDIAGHA', 'DDIAGUN']] == 'Y') == 1:
        if row['DDIAGISC'] == 'Y':
            type_list.append('DDIAGISC')
        elif row['DDIAGHA'] == 'Y':
            type_list.append('DDIAGHA')
        else:
            type_list.append('DDIAGUN')
    else:
        IST_filtered.drop(index, inplace=True)
        
IST_filtered['TYPE'] = type_list
```

```python
IST_filtered.groupby("TYPE").count()
```

```python
IST_filtered[['DDIAGISC', 'DDIAGHA', 'DDIAGUN','TYPE']]
print(IST_filtered['TYPE'].value_counts())
print(IST_filtered['DDIAGISC'].value_counts())
print(IST_filtered['DDIAGHA'].value_counts())
print(IST_filtered['DDIAGUN'].value_counts())
```

```python
IST_filtered.loc[IST_filtered['DDIAGISC'] == 'Y', ['DDIAGISC', 'DDIAGHA', 'DDIAGUN', 'TYPE']]
```

```python
#"DDIAGISC","DDIAGHA","DDIAGUN"
#if all no or more than 1 yes . drop row
#

# columns_included = ["RSLEEP","RHEP24","RASP3","STYPE","DLH14","DMH14","DASP14","FDEAD","DIED","TD"]
columns_included_corrected = ["RSLEEP","RHEP24","RASP3","DLH14","DMH14","DASP14","FDEAD","DIED","TD","TYPE"]

IST_filtered = IST_filtered.loc[:,columns_included_corrected]
IST_filtered = IST_filtered[IST_filtered["RSLEEP"] == "N"]
IST_filtered = IST_filtered[IST_filtered["RHEP24"] == "N"]
IST_filtered = IST_filtered[IST_filtered["RASP3"] == "N"]
# IST_filtered = IST_filtered[(IST_filtered["DNOSTRK"] == "N") | (IST_filtered["DNOSTRK"] == "U")]
# IST_filtered = IST_filtered[(IST_filtered["DNOSTRK"] == "Y")]
IST_filtered = IST_filtered[IST_filtered["DLH14"] == "N"]
IST_filtered = IST_filtered[IST_filtered["FDEAD"] != "U"]

asprin = IST_filtered[IST_filtered["DASP14"] == "Y"]
high_heparin = IST_filtered[IST_filtered["DMH14"] == "Y"]
asprin = asprin[asprin["DMH14"] != "Y"].reset_index(drop=True)
high_heparin = high_heparin[high_heparin["DASP14"] != "Y"].reset_index(drop=True)
asprin["Group"] = "Aspirin"
high_heparin["Group"] = "High_Heparin"
asprin = asprin.drop(columns=["DASP14","DMH14"])
high_heparin = high_heparin.drop(columns=["DASP14","DMH14"])

IST_filtered = pd.concat([asprin, high_heparin], axis=0).reset_index(drop=True)
IST_filtered = IST_filtered.drop(columns=["RSLEEP","RHEP24","RASP3","DLH14"])
IST_filtered = IST_filtered.dropna().reset_index(drop=True)
IST_filtered
```
